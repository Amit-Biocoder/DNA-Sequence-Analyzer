```python
def countnucleotide(L,n):
    c=0
    for i in L:
        if(i==n):
            c+=1
    return c
def complimentaryDNAsequence(L):
    Cdict={
    "A":"T",
    "G":"C",
    "C":"G",
    "T":"A",
    }
    complimentarystrand=[]
    for i in L:
        complimentarystrand.append(Cdict[i])
    sequence=''.join(complimentarystrand)
    return sequence
def mRNA(L):
    for i in range(len(L)) :
        if ( L[i]=="T"):
            L[i]="U"
    mrnasequence=''.join(L)
    return mrnasequence
def longest_orfsequence(L):
    start="AUG"
    stops={"UAA","UAG","UGA"}

    longest=[]
    current=[]
    in_orf=False

    for codon in L :
        if not in_orf:
            if codon==start:
                current.append(start)
                in_orf=True
        else:
            current.append(codon)
            if codon in stops:
                if len(current)>len(longest):
                    longest=current.copy()
                in_orf=False
                current=[]

    return longest
    
def aminoacids(L):
    proteindict={
        "UUU":"Phenylalanine",
        "UUC":"Phenylalanine",
        "UUA":"Leucine",
        "UUG":"Leucine",
        "CUU":"Leucine",
        "CUC":"Leucine",
        "CUA":"Leucine",
        "CUG":"Leucine",
        "AUU":"Isoleucine",
        "AUC":"Isoleucine",
        "AUA":"Isoleucine",
        "AUG":"Methionine",
	    "ACU":"Threonine",
        "ACC":"Threonine",
        "ACA":"Threonine",
        "ACG":"Threonine",
        "AAU":"Asparagine",
        "AAC":"Asparagine",
        "AAA":"Lysine",
        "AAG":"Lysine",
	    "AGU":"Serine",
        "AGC":"Serine",
        "AGA":"Arginine",
        "AGG":"Arginine",	
        "GUU":"Valine",
        "GUC":"Valine",
        "GUA":"Valine",
        "GCU":"Alanine",
        "GCC":"Alanine",
        "GCA":"Alanine",
        "GCG":"Alanine",
        "GAU":"Asparticacid",
        "GAC":"Asparticacid",
        "GAA":"Glutamicacid",
        "GAG":"Glutamicacid",
        "GGU":"Glycine",
        "GGC":"Glycine",
        "GGA":"Glycine",
        "GGG":"Glycine",
        "UCU":"Serine",
        "UCC":"Serine",
        "UCA":"Serine",
        "UCG":"Serine",
        "UAU":"Tyrosine",
        "UAC":"Tyrosine",
        "UAA":"Ochre(Stop)",
        "UAG":"Amber(Stop)",
        "UGU":"Cysteine",
        "UGC":"Cysteine",
        "UGA":"Opal(Stop)",
        "UGG":"Tryptophan",
        "CCU":"Proline",
        "CCC":"Proline",
        "CCA":"Proline",
        "CCG":"Proline",
        "CAU":"Histidine",
        "CAC":"Histidine",
        "CAA":"Glutamine",
        "CAG":"Glutamine",	
        "CGU":"Arginine",
        "CGA":"Arginine",
        "CGG":"Arginine",
        "GUG":"Valine",
        "CGC":"Arginine",
        }
    amino=[]
    for i in L:
        amino.append(proteindict[i])
    AAsequence='-'.join(amino)
    return AAsequence
        
    
```

```python
nucleotidesequence=str(input("Enter your Nucleotide Sequence: "))
L=( "A","T","G","C" )
nucleotide_list=list(nucleotidesequence)
for i in nucleotide_list:
    if i not in L:
        print("Your nucleotide sequence is WRONG")
        raise SystemExit()
nucleotide_count=len(nucleotide_list)
print(" The total number of nucleotides in your sequence is ", nucleotide_count ,"bp")
a=countnucleotide(nucleotide_list,"A")
t=countnucleotide(nucleotide_list,"T")
g=countnucleotide(nucleotide_list,"G")
c=countnucleotide(nucleotide_list,"C")
D={
    "Adenosine count": a,
    "Thiamine count": t,
    "Guanine count": g,
    "Cytosine count": c
}
print(D)
# GC content
GC=(g+c)/nucleotide_count
GCpercent=GC*100
r=round(GCpercent,3)
print("The GC content in the sequence is ", r,"%")
# complimentary DNA strand
CDNA=complimentaryDNAsequence(nucleotide_list)
print("The complimentary DNA to the given sequence is ",CDNA)
#Reverse compliment
RDNA=CDNA[::-1]
print("The Reverse compliment is", RDNA)
#mRNA 
mrna=mRNA(nucleotide_list)
print("The mRNA of the given nucleotide is", mrna)
# GET CODONS
x=len(mrna)-(len(mrna)%3)
codon=[]
for i in range(0,x,3):
    codon.append(mrna[i:i+3])
codonsequence='-'.join(codon)
print("The codons of the given nucleotide sequence will be", codonsequence)
#LONGEST ORF
ORF=longest_orfsequence(codon)
ORF_sequence='-'.join(ORF)
print("The longest ORF sequence is :",ORF_sequence)
#protein sequence
protein=aminoacids(ORF)
print("The protein sequence is :", protein) 

```
